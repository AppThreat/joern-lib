```python
import asyncio

from joern_lib import client, workspace, graph
from joern_lib.detectors import common as cpg, js
from joern_lib.utils import print_table

joern_host = "http://joern:9000"
cpggen_host = "http://cpggen:7072"
joern_username = "admin"
joern_password  = "admin"
```

## Generate CPG



```python
async def generate_cpg():
    connection = await client.get(joern_host, cpggen_host, joern_username, joern_password)
    await workspace.create_cpg(connection, "pkg:maven/commons-validator/commons-validator@1.7")
asyncio.run(generate_cpg())
```

## Inspect methods



```python
async def inspect_methods():
    connection = await client.get(joern_host, cpggen_host, joern_username, joern_password)
    res = await cpg.list_methods(connection, ".*org.apache.commons.validator.routines.UrlValidator.*")
    print_table(res, title="UrlValidator methods")
asyncio.run(inspect_methods())
```

## torch geometric



```python
import torch

async def convert_pyg():
    connection = await client.get(joern_host, cpggen_host, joern_username, joern_password)
    G = await cpg.get_method(connection, "org.apache.commons.validator.routines.UrlValidator.isValid:boolean(java.lang.String)", as_graph=True)
    pyg_graph = graph.to_pyg(G)
    print (pyg_graph)
asyncio.run(convert_pyg())
```
